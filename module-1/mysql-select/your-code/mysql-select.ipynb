```python
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
import pandas as pd
```

## Challenge 01

```python
connection = 'sqlite:///../data/publications.db'
```

```python
engine = create_engine(connection)
type(engine)
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
authors_table = pd.read_sql_query('SELECT * FROM authors', engine)
authors_table
```

```python
publishers_table = pd.read_sql_query('SELECT * FROM publishers', engine)
publishers_table
```

```python
titles_table = pd.read_sql_query('SELECT * FROM titles', engine)
titles_table
```

```python
query01 = 'SELECT authors.au_id AS "AUTHOR ID", authors.au_lname AS "LAST NAME", authors.au_fname AS "FIRST NAME", titles.title AS "Title", publishers.pub_name AS "publishers" FROM authors INNER JOIN titles ON authors.au_id = titleauthor.au_id INNER JOIN titleauthor ON titleauthor.title_id = titles.title_id INNER JOIN publishers ON publishers.pub_id = titles.pub_id'
```

```python
challenge01_table = pd.read_sql_query(query01, engine)
challenge01_table
```

## Challenge 02

```python

query02 = '''
SELECT authors.au_id AS "AUTHOR ID", authors.au_lname AS "LAST NAME", authors.au_fname AS "FIRST NAME", titles.title AS "Title", publishers.pub_name AS "publishers", COUNT(*) AS "Title Count"
FROM authors 
INNER JOIN titles ON authors.au_id = titleauthor.au_id 
INNER JOIN titleauthor ON titleauthor.title_id = titles.title_id 
INNER JOIN publishers ON publishers.pub_id = titles.pub_id 
GROUP BY authors.au_id, authors.au_lname, authors.au_fname, titles.title, publishers.pub_name
'''
```

```python
challenge02_table = pd.read_sql_query(query02, engine)
challenge02_table
```
